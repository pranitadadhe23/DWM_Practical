```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
```

```python
data = pd.read_csv(R"C:\Users\ASUS\Downloads\Titanic-Dataset.csv")  # Replace with actual path
```

```python
print(data.isnull().sum())  # Count missing values
print(data.info())  # Overview of dataset
```

```python
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.show()
```

```python
data_cleaned = data.dropna()
print(data_cleaned.shape)  # Check new shape
```

```python
data.drop(columns=['Cabin'], inplace=True)  # 'Cabin' has too many NaN values
```

```python
data['Age'].fillna(data['Age'].mean())  # Mean
data['Age'].fillna(data['Age'].median())  # Median
data['Embarked'].fillna(data['Embarked'].mode()[0])  # Mode
```

```python
data.fillna(method='ffill')
```

```python
data.fillna(method='bfill')
```

```python
print(data.isnull().sum())  # Check again to confirm missing values are treated
```

# Advanced Techniquie

```python
#Interpolate Missing Values (Good for Time Series)
data['Age'] = data['Age'].interpolate(method='linear')
```

```python
#Using Scikit-Learn’s SimpleImputer
imputer = SimpleImputer(strategy='mean')
data[['Age']] = imputer.fit_transform(data[['Age']])
```

```python
#Using K-Nearest Neighbors Imputer
knn_imputer = KNNImputer(n_neighbors=5)
data[['Age']] = knn_imputer.fit_transform(data[['Age']])
```

```python
print(data.isnull().sum())  # Check again to confirm missing values are treated
```
